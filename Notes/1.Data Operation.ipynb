#### (1) Implementation of Data Operation

```python
import torch
```

```python
# use tensor to represent array
x=torch.arange(12)
x
```

```python
# use shape to access the shape and the amount of elements in tensor
x.shape
```

```python
# num of elements
x.numel()
```

```python
# reshape
x=x.reshape(3,4)
x
```

```python
# all zero tensor
torch.zeros((2,3,4))
```

```python
# all one tensor
torch.ones((2,3,4))
```

```python
# assign tensor with nested lists
# number of "[]" represents the dimension
# batch size=1
torch.tensor([[[2,1,4,3],[1,2,3,4],[4,3,2,1]]]).shape
```

```python
# calculation
x=torch.tensor([1.0,2,4,8])
y=torch.tensor([2,2,2,2])
x+y,x-y,x*y,x/y,x**y
```

```python
# concat two tensors
x=torch.arange(12,dtype=torch.float32).reshape((3,4))
y=torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((x,y),dim=0),torch.cat((x,y),dim=1)  # dim=0: add on top; dim=1: add side-by-side
```

```python
# logical operation
x==y
```

```python
# sum
x.sum()
```

```python
# broadcasting mechanism
a=torch.arange(3).reshape((3,1))
b=torch.arange(2).reshape((1,2))
a,b
```

Tensor a [3, 1] stretches its column to the right ➔ becomes [3, 2]$$\begin{bmatrix} 0 \\ 1 \\ 2 \end{bmatrix} \xrightarrow{\text{stretched}} \begin{bmatrix} 0 & 0 \\ 1 & 1 \\ 2 & 2 \end{bmatrix}$$Tensor b [1, 2] stretches its row downwards ➔ becomes [3, 2]$$\begin{bmatrix} 0 & 1 \end{bmatrix} \xrightarrow{\text{stretched}} \begin{bmatrix} 0 & 1 \\ 0 & 1 \\ 0 & 1 \end{bmatrix}$$

```python
# broadcasting mechanism
a+b
```

```python
# accessing elements
x[-1],x[1:3]
```

```python
# indexing to write element into matrix
x[1,2]=9
x
```

```python
# assign the same value by indexing
x[0:2,:]=12
x
```

```python
# relocation of memory error
before=id(y)
y=y+x
id(y)==before
```

```python
# in comparison
before=id(x)
x+=y
id(x)==before
```

```python
# inplace execution
z=torch.zeros_like(y)
print('id(z):',id(z))
z[:]=x+y
print('id(z):',id(z))
```

```python
# transfer to numpy datatype
a=x.numpy()
b=torch.tensor(a)
type(a),type(b)
```

#### (2) Data Preprocessing

```python
# create csv
import os

os.makedirs(os.path.join('..','data'),exist_ok=True)
data_file=os.path.join('..','data','house_tiny.csv')
with open(data_file,'w') as f:
    f.write('NumRooms,Alley,Price\n')
    f.write('NA,Pave,127500\n')
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

```python
# load from csv
import pandas as pd
data=pd.read_csv(data_file)
data   # or print(data)
```

```python
# to handle missing values, we need to insert or delete
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean(numeric_only=True))
print(inputs)
```

```python
# Convert categorical variables to one-hot vectors and treat NaNs as a distinct category
inputs=pd.get_dummies(inputs,dummy_na=True).astype(float)
print(inputs)
```

```python
# convert the value type in inputs and outputs into tensor type
import torch
x,y=torch.tensor(inputs.values),torch.tensor(outputs.values)
x,y
```
