### **1. Linear Algebra**

#### (1) Basic Objects
- **Scalar**: single number, e.g. `x = 3`.
- **Vector**: 1-D array, e.g. `x = [1, 2, 3]`; often represents one sample or one feature list.
- **Matrix**: 2-D array, shape `(m, n)`; common for a batch of data or a linear transform.
- **Tensor**: general n-D array. In deep learning, images, batches, and model parameters are usually tensors.

Shape is important: most linear algebra bugs are shape mismatches.

#### (2) Matrix and Vector Operations
- **Elementwise operation**: same-shape tensors compute position by position, such as `A + B`, `A * B`.
- **Dot product**: for vectors `x, y`, `x^T y = sum_i x_i y_i`; measures similarity when vectors are normalized.
- **Matrix-vector product**: `A x` maps a vector to another vector.
- **Matrix-matrix product**: `A B` composes two linear transformations.
- **Transpose**: `A^T` swaps rows and columns.

For multiplication, inner dimensions must match: `(m, n) @ (n, k) -> (m, k)`.

#### (3) Reduction and Axis
A reduction summarizes tensor values, for example `sum`, `mean`, `max`.

- `sum(A)`: sum all elements.
- `sum(A, axis=0)`: reduce rows, keep one value for each column.
- `sum(A, axis=1)`: reduce columns, keep one value for each row.
- `keepdims=True`: keeps the reduced dimension, useful for broadcasting later.

#### (4) Broadcasting
Broadcasting lets tensors with compatible shapes participate in elementwise operations.

Example: if `A.shape = (3, 4)` and `b.shape = (4,)`, then `A + b` adds `b` to every row of `A`.

Rule of thumb: compare shapes from right to left; dimensions must be equal or one of them must be `1`.

#### (5) Norms
Norms measure the size of vectors or matrices.

- **L2 norm**: `||x||_2 = sqrt(sum_i x_i^2)`, common length measure.
- **L1 norm**: `||x||_1 = sum_i |x_i|`, often encourages sparsity.
- **Frobenius norm**: matrix norm, `||A||_F = sqrt(sum_ij A_ij^2)`.

In deep learning, norms appear in regularization, gradient clipping, and loss analysis.

#### (6) Why Linear Algebra Matters in Deep Learning
- A neural network layer often has the form `Y = XW + b`.
- Forward propagation is mostly tensor operations.
- Backpropagation computes gradients with respect to tensors.
- GPUs are efficient because large tensor/matrix operations can be parallelized.

Key habit: always track each tensor's shape before and after an operation.

### **2. Realization of Linear Algebra**

```python
import torch
```

```python
# Scalars are represented by tensors with only one element
x=torch.tensor([3.0])
y=torch.tensor([2.0])

x+y,x*y,x/y,x**y
```

```python
# Vectors can be viewed as lists of scalar values
x=torch.arange(4)
x
```

```python
# Access any element by indexing into the tensor
x[0]
```

```python
# Access the length of a tensor
len(x)
```

```python
# Access the shape of a tensor
x.shape
```

```python
# Create an m x n matrix by specifying the two dimensions m and n
A=torch.arange(20).reshape(5,4)
A
```

```python
# Matrix transpose
A.T
```

```python
# The transpose of a symmetric matrix is equal to itself
B=torch.tensor([[1,2,3],[2,0,4],[3,4,5]])
B==B.T
```

```python
# Just as vectors generalize scalars and matrices generalize vectors, we can build data structures with multiple axes
x=torch.arange(24).reshape(2,3,4)
x
```

```python
# Given any two tensors with the same shape, any elementwise binary operation returns a tensor with the same shape
A=torch.arange(20,dtype=torch.float32).reshape(5,4)
B=A.clone()
A,A+B
```

```python
# Elementwise multiplication of two matrices is called the Hadamard product, denoted by \odot in LaTeX
A*B
```

```python
a=2
x=torch.arange(24).reshape(2,3,4)
a+x,(a*x).shape
```

```python
# Calculate the sum of its elements
x=torch.arange(4,dtype=torch.float32)
x,x.sum()
```

```python
# Sum the elements of a tensor with any shape
A=torch.arange(20*2).reshape(2,5,4)
A.shape, A.sum(),A
```

```python
# Specify the axis for summing a tensor; here A.shape=[2,5,4]
A_SUM_AXIS0=A.sum(axis=0)  # Sum corresponding elements across the batch/pages axis; the pages axis is reduced
A_SUM_AXIS0,A_SUM_AXIS0.shape
```

```python
A_SUM_AXIS1=A.sum(axis=1) # Sum elements in the same column within each page; the rows axis is reduced
A_SUM_AXIS1,A_SUM_AXIS1.shape
```

```python
# Sum along two dimensions
A.sum(axis=[0,1]).shape
```

```python
# A quantity related to summation: the mean
A = torch.arange(20, dtype=torch.float32).reshape(5, 4)
A.mean(), A.sum()/A.numel()
```

```python
# Compute the mean along an axis
A.mean(axis=0),A.sum(axis=0)/A.shape[0]
```

```python
# Keep the axes unchanged when calculating the sum or mean
sum_A=A.sum(axis=1,keepdims=True)
sum_A
```

```python
A/sum_A
```

```python
# Calculate the cumulative sum of A's elements along an axis
A.cumsum(axis=0)
```

```python
# The dot product is the sum of elementwise products at the same positions
y=torch.ones(4,dtype=torch.float32)
x,y,torch.dot(x,y)
```

```python
# Express the dot product of two vectors by first multiplying elementwise and then summing
torch.sum(x*y)
```

```python
# Matrix-vector multiplication
A.shape,x.shape,torch.mv(A,x)
```

```python
# Matrix-matrix multiplication
B=torch.ones(4,3)
torch.mm(A,B)
```

```python
# The L1 norm is the sum of the absolute values of the vector elements
u=torch.tensor([3.0,-4.0])
torch.abs(u).sum()
```

```python
# The L2 norm is the square root of the sum of squared vector elements
u=torch.tensor([3.0,-4.0])
torch.norm(u)
```

```python
# The Frobenius norm is the square root of the sum of squared matrix elements
torch.norm(torch.ones((4,9)))
```
